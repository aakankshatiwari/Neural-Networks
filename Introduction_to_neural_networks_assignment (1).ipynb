```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)
```

```python
# Data Initialization
x1 = 3
x2 = 2
y_actual = 17

# Initialize random weights
np.random.seed(42)
w1 = np.random.uniform(1, 10)
w2 = np.random.uniform(1, 10)
learning_rate = 0.01

# Lists to store error and predictions for visualization
errors = []
predictions = []
```

```python
# Training the model
for epoch in range(20):
    # Feed forward
    y_pred = x1 * w1 + x2 * w2

    # Error Calculation
    error = (y_actual - y_pred) ** 2
    errors.append(error)
    predictions.append(y_pred)

    # Back propagation
    gradient_w1 = -2 * (y_actual - y_pred) * x1
    gradient_w2 = -2 * (y_actual - y_pred) * x2

    # Updated weights
    w1 -= learning_rate * gradient_w1
    w2 -= learning_rate * gradient_w2

    print(f'Epoch {epoch + 1}: y_pred = {y_pred:.2f}, error = {error:.2f}')
```

```python
# Visualization
plt.figure(figsize=(14, 5))
```

```python
# Plot error progression
plt.subplot(1, 2, 1)
plt.plot(errors, label='Error')
plt.title('Error Progression')
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.legend()
```

```python
# Plot predicted y progression
plt.subplot(1, 2, 2)
plt.plot(predictions, label='Predicted y', color='orange')
plt.axhline(y=y_actual, color='r', linestyle='--', label='Actual y')
plt.title('Predicted y Progression')
plt.xlabel('Epochs')
plt.ylabel('Predicted y')
plt.legend()

plt.show()
```
